```python
# Function to calculate simple interest

def simple_interest(p,n,r):
    print("SIMPLE Interest function started")
    time.sleep(2)
    i = (p*n*r)/100
    a = p + i
    print(f"simple interest  :  {i:.2f}")
    print(f"Amount : {a:.2f}")
    
```

```python
%%time
simple_interest(p=5000,n=4,r=6.5)
```

```python
def hypotenuse(a,b):
    print("Hypotenuse Function Started")
    time.sleep(3)
    h = (a**2 + b**2 )**(1/2)
    print(f"Hypotenuse of a sides {a} and {b} is {h}") 
    
```

### Mutithreading






```python
# Create multiple thread 

from threading import Thread
```

```python
%%time

# Create the individual threads and assign target function
th1 = Thread(target=simple_interest,args=(5800,6,7.1))
th2 = Thread(target=hypotenuse,args=(7,8))

# Start the threads
th1.start()
th2.start()

# wait for all threads to finish
th1.join()
th2.join()
```

#### Single functions but calculate multiple values

```python
def square(n):
    time.sleep(1)
    s = n**2
    print(f"Square of number {n} is {s}\n\n")
```

```python
%%time
square(5)
```

```python
## single threaded function

def square_list(a: list[int|float]):
    for i in a:
        square(i)
```

```python
a = [2,3,4,5,6,7]
len(a)
```

```python
%%time
square_list(a)
```

```python
#Multithreading

def multi_squares(s: list[int|float]):
    #creates threads for each value and start
    threads = []
    for i in s:
        th = Thread(target=square,args=(i,))
        th.start()
        threads.append(th)


    #wait for all threads to finish
    for th in threads:
        th.join()
```

```python
a
```

```python
%%time
multi_squares(a)
```

```python
b = list(range(1, 101))
print(b)
```

```python
%%time
multi_squares(b)
```

```python
from urllib.request import urlretrieve
```

```python
def download_file(url: str):
    filename = url.split("/")[-1]
    print(f"File {filename} download started ...")
    urlretrieve(url, filename)
    print(f"File {filename} download complete")
```

```python
# Single threading
def download_multiple_files(urls : list[str]):
    for i in urls:
        download_file(i)
        
```

```python
urls
```

```python
%%time
download_multiple_files(urls)
```

```python
# Multithreading
def download_multithread(urls: list[str]):
    # Create and start threads
    threads = []
    for i in urls:
        th = Thread(target=download_file, args=(i,))
        th.start()
        threads.append(th)

    # Wait for all threads to complete
    for th in threads:
        th.join()
```

```python
%%time
download_multithread(urls)
```
