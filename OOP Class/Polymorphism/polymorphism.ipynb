## Polymorphism In Classes

```python
class India:

    def capital(self):
        print("New delhi is a capital of india")

    def language(self):
        print("Hindi is widely spoken language in india")
```

```python
class USA:

    def capital(self):
        print("Washington DC is a capital of USA")

    def language(self):
        print("English is widely spoken language in USA")
```

```python
class France:

    def capital(self):
        print('Paris is a capital of france')

    def language(self):
        print("French is a widely spoken language in France")
```

```python
c1 = India()
c2 = USA()
c3 = France()
```

```python
print(type(c1))
print(type(c2))
print(type(c3))
```

```python
c1.capital()
```

```python
c2.capital()
```

```python
c3.capital()
```

```python
c3.language()
```

```python
c2.language()
```

```python
for i in (c1,c2,c3):
    print(type(i))
    i.capital()
    i.language()
    print("\n__________________________________________\n")
```

#### Polymorphism for calculating area,perimeter of different shapes

```python
from math import pi
```

```python
pi
```

```python
class Rectangle:

    def __init__(self,width,height):
        self.width = width
        self.height = height

    def perimeter(self):
        return 2*(self.height + self.width)
    
    def area(self):
        return self.width * self.height
```

```python
class Circle:

    def __init__(self,r):
        self.r = r

    def perimeter(self):
        return 2*pi*self.r
    
    def area(self):
        return pi * (self.r**2)
```

```python

s2 = Circle(14)
s3 = Rectangle(30, 20)
s5 = Circle(7)
```

```python
shapes = (s2,s3,s5)
```

```python
for i in shapes:
    print(type(i))
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"area : {i.area():.2f}")
    print("\n________________________________________________\n")
```
